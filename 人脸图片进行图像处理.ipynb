```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 文件夹路径
folder_path = "face4"

# 获取文件夹下的所有图片文件名
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

for file in image_files:
    # 读取图片
    image = cv2.imread(os.path.join(folder_path, file))

    # 添加高斯噪声
    gaussian_noise = np.random.normal(loc=0, scale=50, size=image.shape).astype(np.uint8)
    noisy_image1 = cv2.add(image, gaussian_noise)

    # 添加椒盐噪声
    salt_pepper_noise = np.zeros(image.shape[:2], np.uint8)
    cv2.randu(salt_pepper_noise, 0, 255)
    salt_pepper_noise = salt_pepper_noise < 25
    noisy_image2 = cv2.bitwise_and(image, image, mask=np.uint8(salt_pepper_noise))

    # 高斯噪声应用五种滤波方法
    filtered_image1 = cv2.GaussianBlur(noisy_image1, (5, 5), 0)
    filtered_image2 = cv2.medianBlur(noisy_image1, 5)
    filtered_image3 = cv2.bilateralFilter(noisy_image1, 9, 75, 75)
    filtered_image4 = cv2.blur(noisy_image1, (5, 5))
    filtered_image5 = cv2.boxFilter(noisy_image1, -1, (5, 5))

    #椒盐噪声应用五种滤波方法
    filtered_image6 = cv2.GaussianBlur(noisy_image2, (5, 5), 0)
    filtered_image7 = cv2.medianBlur(noisy_image2, 5)
    filtered_image8 = cv2.bilateralFilter(noisy_image2, 9, 75, 75)
    filtered_image9 = cv2.blur(noisy_image2, (5, 5))
    filtered_image10 = cv2.boxFilter(noisy_image2, -1, (5, 5))
    # 显示处理后的图片
    fig, axes = plt.subplots(2,7, figsize=(15, 10))

    axes[0, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0, 0].set_title("Original Image")

    axes[0, 1].imshow(cv2.cvtColor(noisy_image1, cv2.COLOR_BGR2RGB))
    axes[0, 1].set_title("Noisy Image (Gaussian)")

    axes[0, 2].imshow(cv2.cvtColor(filtered_image1, cv2.COLOR_BGR2RGB))
    axes[0, 2].set_title("Filtered Image (Gaussian)")


    axes[0, 3].imshow(cv2.cvtColor(filtered_image2, cv2.COLOR_BGR2RGB))
    axes[0, 3].set_title("Filtered Image (medianBlur)")

    axes[0, 4].imshow(cv2.cvtColor(filtered_image3, cv2.COLOR_BGR2RGB))
    axes[0, 4].set_title("Filtered Image (bilateralFilter)")

    axes[0, 5].imshow(cv2.cvtColor(filtered_image4, cv2.COLOR_BGR2RGB))
    axes[0, 5].set_title("Filtered Image (blur)")

    axes[0, 6].imshow(cv2.cvtColor(filtered_image5, cv2.COLOR_BGR2RGB))
    axes[0, 6].set_title("Filtered Image (boxFilter)")


    axes[1, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[1, 0].set_title("Original Image")

    axes[1, 1].imshow(cv2.cvtColor(noisy_image2, cv2.COLOR_BGR2RGB))
    axes[1, 1].set_title("Noisy Image (salt_pepper)")

    axes[1, 2].imshow(cv2.cvtColor(filtered_image6, cv2.COLOR_BGR2RGB))
    axes[1, 2].set_title("Filtered Image (Gaussian)")


    axes[1, 3].imshow(cv2.cvtColor(filtered_image7, cv2.COLOR_BGR2RGB))
    axes[1, 3].set_title("Filtered Image (medianBlur)")

    axes[1, 4].imshow(cv2.cvtColor(filtered_image8, cv2.COLOR_BGR2RGB))
    axes[1, 4].set_title("Filtered Image (bilateralFilter)")

    axes[1, 5].imshow(cv2.cvtColor(filtered_image9, cv2.COLOR_BGR2RGB))
    axes[1, 5].set_title("Filtered Image (blur)")

    axes[1, 6].imshow(cv2.cvtColor(filtered_image10, cv2.COLOR_BGR2RGB))
    axes[1, 6].set_title("Filtered Image (boxFilter)")


    axes[1, 6].axis('off')

    plt.tight_layout()
    plt.show()
```
